# Regresion Lineal

La funcion de costo para una regresion lineal regularizada es la siguiente:	
 $\frac{1}{2n} \sum_{i=1}^{n} (h_{0} (x^{(i)}) - y^{(i)})^2 $

### Funcion de costo y de la regresión lineal regularizada

```python
def cost(theta, X, y, lamda):
    n = len(X)
    #print("theta: ", theta)
    predictions = X.dot(theta)
    #print("cost: ", ((n/2) * np.sum(np.square(predictions-y))))
    # return ((1/2*n) * np.sum(np.square(predictions-y))) + ((lamda * np.sum(np.square(theta))) / (2*n))
    return (np.sum(np.square((predictions - y)))/(2 * n)) + (np.sum(np.square(theta)) * lamda / (2 * n))


def gradient_vectorized(X, y, theta, lamda, learning_rate=0.01, iterations=1000):
    N = len(X)
    cost_history = np.zeros(iterations)
    prediction_history = np.zeros(iterations)
    prediction = 0
    for i in range(iterations):
        
        theta = theta - learning_rate*((X.T.dot((prediction - y)))/N) + ((lamda * theta)/N)
        """
        # Prediction values
        prediction = np.dot(X, theta)
        # Difference in real values and predicted values 
        loss = prediction - y 
        # Gradient calculation
        gradient = X.T.dot(loss) / N
        # Change theta values based on the gradient
        theta = theta - learning_rate * gradient
        cost_history[i] = cost(theta, X, y, lamda)
        """
    return theta, cost_history, prediction

def cost_function(X, Y, B):
    m = len(Y)
    J = np.sum((X.dot(B) - Y) ** 2)/(2 * m)
    return J

def batch_gradient_descent(X, Y, B, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)

    for iteration in range(iterations):
        #print(iteration)
        # Hypothesis Values
        h = X.dot(B)
        # Difference b/w Hypothesis and Actual Y
        loss = h - Y
        # Gradient Calculation
        gradient = X.T.dot(loss) / m
        # Changing Values of B using Gradient
        B = B - alpha * gradient
        # New Cost Value
        cost = cost_function(X, Y, B)
        cost_history[iteration] = cost

    return B, cost_history, h
```

## Elección de datos

```python
#X = 2 * np.random.rand(100, 1)
#y = 4 + 3 * X + np.random.randn(100, 1)
#sc = StandardScaler()
#X = sc.fit_transform(X)
df['Admit'] = df['Chance of Admit '] * 100
X = np.array(df['GRE Score'])

y = np.array(df['Admit'])
y = y.astype(int)
#print(X)
#print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Initial Coefficients
B = np.zeros((2, 1))
#B = np.random.randn(2, 1)
print(B)
alpha = 0.000005
iter_ = 1000

X_b = np.c_[np.ones((len(X_train), 1)), X_train]

newB, cost_history, y_predict = batch_gradient_descent(X_b, y_train, B, alpha, iter_)
```
