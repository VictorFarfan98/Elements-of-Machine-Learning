# Clean the Dataset

## Drop columns where all rows are NA

```python
la_df = la_df.drop(columns=['Type', 'Grade','Environment','Quality','Condition','Appearance','Storage','Crop','Trans Mode'])
ny_df = ny_df.drop(columns=['Type', 'Grade','Environment','Quality','Condition','Appearance','Storage','Crop','Trans Mode'])
```

## Drop columns where mayority of observations are NA

* Drop *Sub Variety* column because almost 90% of observations are NA and we still have the *Variety* column without any NA observation
* Drop *color* column because almost 90% of observations are NA and I don't think that the pumpkin's color is too important

```python
la_df = la_df.drop(columns=['Sub Variety'])
ny_df = ny_df.drop(columns=['Sub Variety'])
```
